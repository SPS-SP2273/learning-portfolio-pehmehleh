# 1 A simple plot

## 1.1 Letâ€™s look at some code.

```python
x = [0, 1, 2, 3, 4, 5, 6]
y_1 = [0, 2, 4, 6, 8, 10, 12]
y_2 = [0, 3, 6, 9, 12, 15, 18]

#plotting
plt.plot(x,y_1, label="Y values", color = "purple", linestyle = "dashed") #this is the line
plt.xlabel("x-values")
plt.ylabel("y-values")
plt.title("X vs Y")
plt.grid(alpha = 0.3) #this determines transparency of the gridlines behind
plt.legend(loc = "center right") #location of legend. CENTER NOT CENTRE

#'bottom right' is not a valid value for loc
#supported values are 'best', 'upper right', 'upper left', 'lower left', 'lower right', 'right', 'center left', 'center right', 'lower center', 'upper center', 'center'
```

### Things to note

- **abbreviations**
- color: c (DONT TYPE AS COLOUR!!!)
- linestyle: ls
- linewidth: lw
- will see output even if omit plt.show() because jupyter is interactive environment but good to include
- plotting more than one line is easy, just put plt.plot(x,y_1,y_2) - gives 3 lines
- can split arguments into separate lines. order of specifying keyword arguments does not matter
- #supported values for legend location are 'best', 'upper right', 'upper left', 'lower left', 'lower right', 'right', 'center left', 'center right', 'lower center', 'upper center', 'center'
- #supported values for linestyle are '-', '--', '-.', ':', 'None', ' ', '', 'solid', 'dashed', 'dashdot', 'dotted'

```python
x = [0, 1, 2, 3, 4, 5, 6]
y_1 = [0, 2, 4, 6, 8, 10, 12]
y_2 = [0, 3, 6, 9, 12, 15, 18]

#plotting
plt.plot(x,y_1, label="Y values", c = "pink", ls = "-", lw = 3) #this is the line
plt.xlabel("x-values")
plt.ylabel("y-values")
plt.title("X vs Y")
plt.grid(alpha = 0.3) #this determines transparency of the gridlines behind
plt.legend(loc = "center right") #location of legend. CENTER NOT CENTRE

#'bottom right' is not a valid value for loc
#supported values for legend location are 'best', 'upper right', 'upper left', 'lower left', 'lower right', 'right', 'center left', 'center right', 'lower center', 'upper center', 'center'
#supported values for linestyle are '-', '--', '-.', ':', 'None', ' ', '', 'solid', 'dashed', 'dashdot', 'dotted'
```

```python
#the two plots below produce the same result. difference: abbreviation
plt.plot(x, y_1, color='red', linestyle='dashed', linewidth=2)
plt.show()
```

```python
plt.plot(x, y_1, c='red', ls='dashed', lw=2)
plt.show() #GOOD TO INCLUDE
```

```python
x = [0, 1, 2, 3, 4, 5, 6]
y_1 = [0, 2, 4, 6, 8, 10, 12]
y_2 = [0, 3, 6, 9, 12, 15, 18]

#plotting more than 1 line
plt.plot(x,y_1,y_2, #splitting arguments here can increase readability. order does not matter
         label="Y values",
         ls = "-",
         lw = 3) #default colours are orange and blue
plt.xlabel("x-values")
plt.ylabel("y-values")
plt.title("X vs Y")
plt.grid(alpha = 0.3) #this determines transparency of the gridlines behind
plt.legend(loc = "center right") #location of legend. CENTER NOT CENTRE
```

## 1.2 Adding another plot (color NOT colour!!!)

```python
plt.plot(x,y_1, label="Y values", color = "purple", linestyle = "dashed") #line 1
plt.plot(x,y_2, label = "Y2 values", color = "green", linestyle = "--") #line 2
plt.xlabel("x-values") #axis title
plt.ylabel("y-values") #acis title
plt.title("X vs Y")
plt.grid(alpha = 0.3) #this determines transparency of the gridlines behind
plt.legend(loc = "lower right") #location of legend. CENTER NOT CENTRE
plt.show()
```

## 1.3 Yet another plot but with error bars using errorbar()

```python
y_3 = [0, 4, 8, 12, 16, 20, 24]
x_error, y_error = .1, 0.75 #these are constant errors for all the points. can put in list of errors so each will have different length

plt.plot(x,y_1, label="Y values", color = "purple", linestyle = "dashed") #line 1
plt.plot(x,y_2, label = "Y2 values", color = "pink", linestyle = "-") #line 2
plt.xlabel("x-values") #axis title
plt.ylabel("y-values") #acis title
plt.title("X vs Y")
plt.grid(alpha = 0.3) #this determines transparency of the gridlines behind
plt.legend(loc = "lower right") #location of legend. CENTER NOT CENTRE
#error bars
plt.errorbar(x, y_3, xerr = x_error, yerr = y_error, label = "Y3 with errors", color="green")

plt.show()
```

# 2 Better with NumPy

- easier to use numpy arrays instead of python lists
- **one main benefit: generate data-related mathematical fxns earily, eg. x^2 and sinx**
- make np arrays from lists first

```python
import numpy

# Some data for plotting
x = [0, 1, 2, 3, 4, 5, 6]
y_1 = [0, 2, 4, 6, 8, 10, 12]
y_2 = [0, 3, 6, 9, 12, 15, 18]

np_x = numpy.array(x)
np_y_1 = numpy.array(y_1)
np_y_2 = numpy.array(y_2)

plt.plot(np_x, np_y_1, color='red', linestyle='dashed', label='Y values')
plt.plot(np_x, np_y_2, color='blue', label='Y2 values')
```

## 2.1 Adding mathematical functions

```python
x = numpy.array([0,1,2,3,4,5,6])
x2 = x ** 2 #square
sinx = numpy.sin(x)

plt.plot(x,x2, label = "x^2", color="red", ls="--")
plt.plot(x,sinx, label = "sin(x)", color = "blue", ls = "-")
plt.legend()
plt.show()

#DOES NOT LOOK GOOD BECAUSE SIN X IS BTW +- 1.
```
